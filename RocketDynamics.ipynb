```python
from math import sin, cos, log, ceil
import numpy
import matplotlib.pyplot as plt
%matplotlib inline
from matplotlib import rcParams
rcParams['font.family'] = 'serif'
rcParams['font.size'] = 16
```

```python
# model parameters:
g = 9.81     # gravity in m s^{-2}
v_e = 325.0    # exhaust velocity, m s^{-1}
m_s = 50.0   # mass of the rocket shell, kg
C_D = 0.15   # drag coefficient
r = 0.50      # maximum cross sectional radius, m
rho = 1.091  # average air density, kg m^{-3}
A = numpy.pi*(r**2) # Area of Cross Section, m^{2}

### set initial conditions ###
h0 = 0.0       # initial height, m
v0 = 0.0       # initial velocity, m s^{-1}
m_p0 = 100.0   # initial mass of the rocket propellant, kg
```

```python
def f(u):
    """Returns the right-hand side of the rocket system of equations.
    
    Parameters
    ----------
    u : array of float
        array containing the solution at time n.
        
    Returns
    -------
    dudt : array of float
        array containing the RHS given u.
    """ 
    
    h = u[0]
    v = u[1]
    return numpy.array([v,
                      -g + (mpdotval*v_e/(m_s+m_p[n])) - \
                       ((0.5*rho*v*abs(v)*A*C_D)/(m_s+m_p[n]))])
```

```python
def euler_step(u, f, dt):
    """Returns the solution at the next time-step using Euler's method.
    
    Parameters
    ----------
    u : array of float
        solution at the previous time-step.
    f : function
        function to compute the right hand-side of the system of equation.
    dt : float
        time-increment.
    
    Returns
    -------
    u_n_plus_1 : array of float
        approximate solution at the next time step.
    """
    
    return u + dt * f(u)
```

```python
T = 40.0                           # final time
dt = 0.1                           # time increment
N = int(T/dt) + 1                  # number of time-steps
t = numpy.linspace(0.0, T, N)      # time discretization

m_p = numpy.empty_like(t,dtype=numpy.ndarray)
mpdot = numpy.zeros_like(t,dtype=numpy.ndarray)
mpdot[0:(5/dt)] = 20.0
mpdotint = 100.0*numpy.ones_like(t,dtype=numpy.ndarray)
mpdotint[0:(5/dt)] = 20.0*t[0:(5/dt)]  # setting the constant burning rate for 5 sec

# initialize the array containing the solution for each time-step
u = numpy.empty((N, 2))
u[0] = numpy.array([h0,v0])# fill 1st element with initial values

# time loop - Euler method
for n in range(N-1):
    mpdotval = mpdot[n]
    m_p[n] = m_p0 - mpdotint[n]
    u[n+1] = euler_step(u[n], f, dt)
```

```python
h = u[:,0]
v = u[:,1]

# visualization of the variables
plt.figure(figsize=(8,6))
plt.grid(True)
plt.xlabel(r't', fontsize=18)
plt.ylabel(r'h', fontsize=18)
plt.title('Height (m) vs. Time (s)', fontsize=18)
plt.plot(t,h, 'k-', lw=2);

plt.figure(figsize=(8,6))
plt.grid(True)
plt.xlabel(r't', fontsize=18)
plt.ylabel(r'v', fontsize=18)
plt.title('Velocity (m/s) vs. Time (s)', fontsize=18)
plt.plot(t,v, 'k-', lw=2);

plt.figure(figsize=(8,6))
plt.grid(True)
plt.xlabel(r't', fontsize=18)
plt.ylabel(r'm_p', fontsize=18)
plt.title('Propellant Mass (kg) vs. Time (s)', fontsize=18)
plt.plot(t,m_p, 'k-', lw=2);
```
