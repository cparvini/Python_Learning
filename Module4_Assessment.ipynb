```python
%matplotlib inline
import numpy as np
import sympy
import pylab
from scipy import interpolate
import random
import matplotlib.pyplot as plt
import matplotlib.cm as cm
sympy.init_printing()

n = 192
Du, Dv, F, k = 0.00016, 0.00008, 0.035, 0.065 # Bacteria 1 
dh = 5./(n-1)
T = 8005
dt = .9 * (dh**2) / (4*max(Du,Dv))
nt = int(T/dt)
t = np.arange(0,T,dt)

uvinitial = np.load('./uvinitial.npz')
Ui = uvinitial['U']
Vi = uvinitial['V']

Un = np.zeros((n,n,nt+1))
Vn = np.zeros((n,n,nt+1))

for i in range(0,nt):    
    Un[:,:,i] = Ui.copy()
    Vn[:,:,i] = Vi.copy()
    
    # Next time step for the internal grid    
    Un[1:-1,1:-1,i+1] = (Du*dt/(dh**2)) * (Un[2:,1:-1,i]+Un[0:-2,1:-1,i]+Un[1:-1,2:,i]+Un[1:-1,0:-2,i]) +\
        F*dt + Un[1:-1,1:-1,i] * (F*dt + (4*Du*dt/(dh**2)) - dt*Vn[1:-1,1:-1,i]**2 + 1)
    
    Vn[1:-1,1:-1,i+1] = (Dv*dt/(dh**2)) * (Vn[2:,1:-1,i]+Vn[0:-2,1:-1,i]+Vn[1:-1,2:,i]+Vn[1:-1,0:-2,i]) +\
        Vn[1:-1,1:-1,i] * (1 + Un[1:-1,1:-1,i]*Vn[1:-1,1:-1,i] - (F + k)*dt - (4*Dv*dt/(dh**2)))
    
    # Neumann along the boundaries
    Un[1:-1,0,i+1] = Un[1:-1,1,i+1]
    Un[1:-1,-1,i+1] = Un[1:-1,-2,i+1]
    Vn[0,1:-1,i+1] = Vn[1,1:-1,i+1]
    Vn[-1,1:-1,i+1] = Vn[-2,1:-1,i+1]
    
    # Neumann at the corners
    Un[-1,0,i+1] = Un[-2,1,i+1] # Bottom Right
    Un[0,0,i+1] = Un[1,1,i+1] # Bottom Left
    Un[-1,-1,i+1] = Un[-2,-2,i+1] # Top Right
    Un[0,-1,i+1] = Un[1,-2,i+1] # Top Left
    
    Vn[-1,0,i+1] = Vn[-2,1,i+1] # Bottom Right
    Vn[0,0,i+1] = Vn[1,1,i+1] # Bottom Left
    Vn[-1,-1,i+1] = Vn[-2,-2,i+1] # Top Right
    Vn[0,-1,i+1] = Vn[1,-2,i+1] # Top Left

fig = plt.figure(figsize=(8,5))
plt.subplot(121)
plt.imshow(Un[:,:,-1],cmap=cm.RdBu)
plt.xticks([]), plt.yticks([]);
plt.subplot(122)
plt.imshow(Vn[:,:,-1],cmap=cm.RdBu)
plt.xticks([]), plt.yticks([]);
```
